```python
# 1.  Function definition
def hello():
    print('Muhammad ali Shahid')

hello() # Function calling
```

```python

# 2. Function with parameter
def hello2(name):
    print(f'hello {name}')

hello2('Muhammad ali ') # Function calling
```

```python

# 3. Function with parameter and return value
def hello3(name):
    return f'hello {name}'

res = hello3('Muhammad ali ') # Function calling.
print(res)
```

```python
# ---------------------------------------------------------
# Case no: 1
def sum1(a, b): # a and b are parameters.
    c = a + b
    return c # returned value

# function calling
x = sum1(10, 15) # 10 and 15 are arguments.
print(x)

```

```python
# Case no: 2
def sum2(a = 20, b = 15): # Default parameters.
    x = a + b
    return x # returned value

answer  = sum2()
print(answer)
```

```python
# Case no: 3
answer = sum2(a = 5, b = 26) # Keyword Arguments
print(answer)
```

```python
# 4. Lamda function
add = lambda x, y : x + y

result = add(22, 3.7)
print(result)
```

```python
# 5. Function with unlimited auguments
def sum(*a):
    sum = 0  # 10, 89, 10, 89, 78, 14, 15
    for i in a:
        sum = i + sum
    
    return sum

sum(10, 89, 10, 89, 78, 14, 15,100)
```

```python

# 6. Function with unlimited keyword auguments
def sum(**a):
    print(a)

sum(a = 56, b = 35, c = 67, d = 10, e = 67)
```

```python

# 7. Recursive functions
def name(a):
    print('----------------')
    print(a)
    name(a)

name('Muhammad ali')
```
